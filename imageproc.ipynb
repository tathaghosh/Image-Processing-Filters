##Assignment1-Luminance Operations(Contrast)


```
%matplotlib inline
from PIL import Image,ImageEnhance
from matplotlib import pyplot as plt
```

```
def main():
  img=Image.open("1st_problem.jpg")
  en=ImageEnhance.Contrast(img)
  img2=en.enhance(0.0)
  img3=en.enhance(0.5)
  img4=en.enhance(2.0)
   
  
  plt.subplot(1, 4, 1)
  plt.imshow(img)
  plt.title('Original Image')
  
  plt.subplot(1, 4, 2)
  plt.imshow(img2)
  plt.title('Zero Contrast')


  plt.subplot(1, 4, 3)
  plt.imshow(img3)
  plt.title('Lower Contrast')

  plt.subplot(1, 4, 4)
  plt.imshow(img4)
  plt.title('Higher Contrast')  
  
  plt.show() 
```

```
if __name__ == "__main__":
    main()
```

##**Assignment2-Linear filtering**

```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
def main():
    
    img = cv2.imread("2nd_problem.jpg", 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    
    #Guassian blur    
    g = np.array(([1, 2, 1], [2, 4, 2], [1, 2, 1]), np.float32)/16

    #Edge filter
    e = np.array(([-1, -1, -1], [-1, 8, -1], [-1, -1, -1]), np.float32)
    
    #Sharpen filter
    s = np.array(([0, -1, 0], [-1, 5, -1], [0, -1, 0]), np.float32)

        

    gaussian = cv2.filter2D(img, -1, g)
    edge= cv2.filter2D(img, -1, e)
    sharpen= cv2.filter2D(img, -1, s)
    
    titles = ['Original Image', 'Sharpen', 
              'Edge', 'Gaussian Blur']

    outputs = [img, sharpen, edge, gaussian]
    
    
    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.imshow(outputs[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()
```

```
if __name__ == "__main__":
    main()
```

##Assignment 3

```
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
```

```
def main():
  img = cv2.imread("3rd_problem.jpg", 1)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  linear=cv2.resize( img, None, fx=1, fy=1, interpolation=cv2.INTER_LINEAR)
  up=cv2.resize( img, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC )
  down=cv2.resize( img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA )
  
  
  
  plt.subplot(1, 3, 1)
  plt.imshow(linear)
  plt.title('Original Image')
  
  plt.subplot(1, 3, 2)
  plt.imshow(up)
  plt.title('Up Image')


  plt.subplot(1, 3, 3)
  plt.imshow(down)
  plt.title('Down Image')  
  
  plt.show() 
  
  
  #plt.imshow(img)
  #plt.show()
```

```
if __name__ == "__main__":
    main()
```

```
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
```

```
def main():
    
        
    img1 = cv2.imread("4th_background.jpg", 1)
    img2 = cv2.imread("4th_foreground.jpg", 1)
    img3 = cv2.imread("4th_mask.jpg", 1)
    
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
    
    img4 = cv2.bitwise_or(img1, img3)
    img5 = cv2.bitwise_and(img2, img4)
    
    plt.imshow(img5)
    

    plt.show()  
```
